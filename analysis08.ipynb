```python
## code for normalization 
from tqdm import tqdm
def normalize(array):
    
    def mean(array):
        add=0
        for nums in array: add += nums
        mean_val = add / len(array)
        return mean_val
    
    def std_dev(array, mean_val):
        summ=0
        for nums in array:
            numer = (nums - mean_val)**2
            summ+=numer
        std_dev_val = (summ / len(array)) ** 0.5
        return std_dev_val
    
    mean_val = mean(array)
    std_dev_val = std_dev(array, mean_val)
    norm_val=[]
    for nums in array:
        z = (nums - mean_val) / std_dev_val
        z_new = float("{:.4f}".format(z))
        norm_val.append(z_new)
    return norm_val
cols = final.columns
time=0
for col in tqdm(cols):
    normalized = normalize(final[col].to_list())
    if time==0:
        data = pd.DataFrame(normalized, columns=[col])
        time+=1
    else:
        data2=pd.DataFrame(normalized, columns=[col])
        data = pd.concat([data, data2], axis=1)
print(data)     
```
